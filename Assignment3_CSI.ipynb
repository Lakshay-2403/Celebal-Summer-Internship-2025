```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_vals = [0, 1, 2, 3, 4]
y_vals = [0, 2, 4, 6, 8]

plt.figure(figsize=(10, 6), dpi=120)

plt.plot(x_vals, y_vals, 'g^--', label='y = 2x')

x_interval = np.arange(0, 4.5, 0.5)

plt.plot(x_interval[:6], x_interval[:6]**2, 'orange', label='y = x^2')
plt.plot(x_interval[5:], x_interval[5:]**2, color='orange', linestyle='--')

plt.title('Sample Plot: Line and Parabola', fontdict={'fontname': 'Arial', 'fontsize': 18})
plt.xlabel('X Value')
plt.ylabel('Y Value')

plt.xticks([0, 1, 2, 3, 4, 5])

plt.legend()

plt.savefig('output_graph.png', dpi=300)

plt.show()
```

```python
categories = ['X', 'Y', 'Z']
amounts = [1, 4, 2]

plt.figure(figsize=(6, 4), dpi=120)

graph_bars = plt.bar(categories, amounts)

hatch_patterns = ['|', '-', '+']
for bar in graph_bars:
    bar.set_hatch(hatch_patterns.pop(0))

plt.savefig('custom_barchart.png', dpi=300)

plt.show()
```

```python
data = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(9, 6))

plt.title('Fuel Costs Over Time (in USD)', fontdict={'fontweight':'light', 'fontsize': 16})

plt.plot(data.Year, data.USA, 'm--^', label='USA')
plt.plot(data.Year, data.Canada, 'c--o', label='Canada')
plt.plot(data.Year, data['South Korea'], 'b--s', label='South Korea')
plt.plot(data.Year, data.Australia, 'g--D', label='Australia')

selected_countries = ['USA', 'Canada', 'South Korea', 'Australia']
for country in selected_countries:
    plt.plot(data.Year, data[country], linestyle='--', marker='x')

plt.xticks(data.Year[::4].tolist() + [2010])

plt.xlabel('Year')
plt.ylabel('Price in USD')

plt.legend()

plt.savefig('fuel_cost_chart.png', dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.hist(data['USA'], bins=10, alpha=0.5, label='USA', color='blue', edgecolor='black')
plt.hist(data['Canada'], bins=10, alpha=0.5, label='Canada', color='red', edgecolor='black')
plt.hist(data['South Korea'], bins=10, alpha=0.5, label='South Korea', color='green', edgecolor='black')
plt.hist(data['Australia'], bins=10, alpha=0.5, label='Australia', color='yellow', edgecolor='black')

plt.xlabel('Gas Price (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Gas Prices by Country')

plt.legend()

plt.savefig('gas_prices_histogram.png', dpi=300)

plt.show()
```

```python
most_recent_year = data['Year'].max()

latest_year_data = data[data['Year'] == most_recent_year]

plt.figure(figsize=(8, 8))

plt.pie(latest_year_data[['USA', 'Canada', 'South Korea', 'Australia']].values.flatten(),
        labels=['USA', 'Canada', 'South Korea', 'Australia'],
        autopct='%1.1f%%',
        startangle=140,
        colors=['blue', 'red', 'green', 'yellow'],
        wedgeprops={'edgecolor': 'black'})

plt.title(f'Gas Prices Distribution in {most_recent_year} (USD)')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.boxplot([data['USA'], data['Canada'], data['South Korea'], data['Australia']], 
            labels=['USA', 'Canada', 'South Korea', 'Australia'], 
            patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            medianprops=dict(color='red'))

plt.title('Gas Prices Box and Whisker Plot')
plt.ylabel('Gas Price (USD)')

plt.show()
```
